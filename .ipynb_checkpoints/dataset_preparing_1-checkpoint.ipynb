### Removing columns

First, we'll remove columns which won't be necessary in further analysis. We will remove the code columns corresponding to categorical text columns (e.g. country and country_txt columns, remove country), because we want to use the categorical datatype with the text columns and reduce the data amount as much as possible.

```python
# We don't need the eventid column as it essentially represents the incident date
data = data.drop('eventid', axis=1)

# It will be easier to work with string values => remove country, region and alternative
data = data.drop('country', axis=1)
data = data.drop('region', axis=1)
data = data.drop('alternative', axis=1)

# We will not look at the specific descriptions of the incidents or their locations => remove the summary and location columns
data = data.drop('summary', axis=1)
data = data.drop('location', axis=1)

# Remove other non-text duplicate columns
data = data.drop('attacktype1', axis=1)
data = data.drop('attacktype2', axis=1)
data = data.drop('attacktype3', axis=1)
data = data.drop('targtype1', axis=1)
data = data.drop('targsubtype1', axis=1)
data = data.drop('natlty1', axis=1)
data = data.drop('targtype2', axis=1)
data = data.drop('targsubtype2', axis=1)
data = data.drop('natlty2', axis=1)
data = data.drop('targtype3', axis=1)
data = data.drop('targsubtype3', axis=1)
data = data.drop('natlty3', axis=1)
```

### Reducing memory usage I: int and float

In order to reduce memory usage, we will downcast the int and float types.

```python
for col in data.select_dtypes(include=['int64']):
    data[col] = pd.to_numeric(data[col], downcast ='integer')
    
for col in data.select_dtypes(include=['float']):
    data[col] = pd.to_numeric(data[col], downcast ='float')

print(data.info(memory_usage='deep'))
```

### iyear, imonth, iday and approxdate

```python
# There are 0s in imonth and iday columns
print('Days with value 0:', data.iday.value_counts()[0])
print('Months with value 0:', data.imonth.value_counts()[0])

# According to the documentation, for attacks that took place between 1970 and 2011, if the exact month of the event is unknown, 
# then this is recorded as "0"
print('Days 0 after 2011:', len(data[(data.iday == 0) & (data.iyear > 2011)]))
print('Months 0 after 2011:', len(data[(data.imonth == 0) & (data.iyear > 2011)]))

# If the day is missing then so is always the month, but not the other way around
print('Day 0 or month 0:', len(data[(data.imonth == 0) | (data.iday == 0)]))
print('Day 0 and month NOT 0:', len(data[(data.imonth == 0) & (data.iday != 0)]))
print('Day NOT 0 and month 0:', len(data[(data.imonth != 0) & (data.iday == 0)]))
```

```python
# We have an approxdate for every missing date (day or month 0)
print(len(data[((data.imonth == 0) | (data.iday == 0)) & (data.approxdate != "")]))
print(len(data[((data.imonth == 0) | (data.iday == 0)) & (data.approxdate == "")]))
```

Whenever the exact date of the incident is not known or remains unclear, this field 'approxdate' is used to record the approximate date of the incident. To include as much information as possible, we will create our own column 'date' which contains either the exact or the approximate date. 
We will look at the 'approxdate' values where the data for the incident month or day is missing (i.e. 'imonth' or 'iday' column values are 0).

```python
data[((data.imonth == 0) | (data.iday == 0)) & (data.approxdate != "")].approxdate.unique()
```

From this list, we'll consider three date formats (dd/mm/yyyy, mm/dd/yyyy, yyyy-mm-dd HH:MM:SS) and two regex forms for the dates: 'Month Day, Year' (e.g. 'May 10, 2006') and 'Month Day1-Day2, Year' (e.g. 'October 28-29, 2010'). The precise date of the latter will be calculated as the average of Day1 and Day2. We will not consider durations where there are two months (e.g. 'March 31-April1, 1997').

```python
import datetime as dt
import re
import math

formats = ["%d/%m/%Y", "%m/%d/%Y","%Y-%m-%d %H:%M:%S"]
month_regex = "(January|February|March|April|May|June|July|August|September|October|November|December)"
regex = [month_regex + " [0-9]{1,2}-[0-9]{1,2}, [0-9]{4}",
         month_regex + " [0-9]{1,2}, [0-9]{4}"]
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
data['date'] = np.nan
data.date = pd.to_datetime(data.date, format = "%d/%m/%Y")

for idx, row in data.iterrows():
    corrected = False
    if idx % 1000 == 0:
        print(idx)
        
    if row.imonth != 0 and row.iday != 0:
        data['date'][idx] = dt.date(year=row.iyear, month=row.imonth, day=row.iday)
        continue 
        
    if row.approxdate != "" and (row.imonth == 0 or row.iday == 0):
        for fmt in formats:
            try:
                date = dt.datetime.strptime(row.approxdate, fmt).date()
                corrected = True
                data['date'][idx] = date
                break
            except ValueError:
                continue
                
        if not corrected:
            for i in range(len(regex)):
                groups = re.match(regex[i], row.approxdate)
                if groups != None and i == 0:
                    pieces = row.approxdate.split(" ")
                    days = pieces[1].split("-")
                    day = round((int(days[0]) + int(days[1][:-1])) / 2)
                    month = months.index(pieces[0]) + 1
                    year = int(pieces[2])
                    data['date'][idx] = dt.date(year=year, month=month, day=day)
                    break
                elif groups != None and i == 1:
                    pieces = row.approxdate.split(" ")
                    day = int(pieces[1][:-1])
                    month = months.index(pieces[0]) + 1
                    year = int(pieces[2])
                    data['date'][idx] = dt.date(year=year, month=month, day=day)
                    break
```

### Replacing text values

We want to replace duplicate string values in the columns with the correct value (i.e. if we have values 'new york' and 'NeW york' in the city column then we want to replace both with 'New York'). For this, we check if there are only upper- or lowercase differences between values in the columns. We will only keep the title-case values.

```python
def clean_case_insensitive(data, colname):
    lst = data[colname].unique()
    lst_lower = list(map(lambda s: s.lower(), lst))
    set_ = set(lst_lower)
    has_dupl = len(lst) != len(set_)
    print("'{}' has duplicates:{}\n".format(colname, has_dupl))
    if has_dupl:
        duplicates = [lst[i] for i in range(len(lst)) if lst_lower.count(lst_lower[i]) > 1 and lst[i] != lst[i].title()]
        print(duplicates)
        for d in duplicates:
            data = data.replace(d, d.title())
    
    return data
```

First, we clean the country_txt, region_txt and provstate columns.

```python
data = clean_case_insensitive(data, 'country_txt')
data = clean_case_insensitive(data, 'region_txt')
data = clean_case_insensitive(data, 'provstate')
```

We can see that there are more duplicate values in the column: 'Pattani (province)', 'Pattani province', 'Narathiwat (province)' and 'Narathiwat province'. We'll replace those too (now title-case).

```python
# Check number of instances with these values
print(len(data[data.provstate == 'Pattani (Province)']), len(data[data.provstate == 'Pattani Province']))
print(len(data[data.provstate == 'Narathiwat (Province)']), len(data[data.provstate == 'Narathiwat Province']))
```

```python
# Replace
data = data.replace('Pattani (Province)', 'Pattani Province')
data = data.replace('Narathiwat (Province)', 'Narathiwat Province')
```

### Reducing memory usage II: object type

We want to convert the object types to categorical types so the memory usage would be optimal. We could not do this before as we needed to parse these text values (pandas internally represents categorical values as integers).

```python
categorical_cols = []
for col in data.select_dtypes(include=['object']):
    num_unique_values = len(data[col].unique())
    num_total_values = len(data[col])
    prop = num_unique_values / num_total_values
    if prop < 0.5:
        if col != 'approxdate':  # This is not a categorical value
            categorical_cols.append(col)
        print(col, num_unique_values, str(round(prop*100))+'%')
```

```python
for cols in categorical_cols:
    data[cols] = data[cols].astype('category')

print(data.info(memory_usage='deep'))
```
