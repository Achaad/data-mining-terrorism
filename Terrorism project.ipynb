```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("globalterrorismdb_0718dist.csv",encoding="iso-8859-1",low_memory=False)
```

```python
data.head()
```

```python
# Setting that allows notebook to display all columns
pd.options.display.max_columns = None
# Describes the data in the dataset
data.describe(include="all")
```

```python
# Plots the iyear to shaw the distribution of incidents by years
plt.hist(data.iyear)
plt.show()
```

```python
# Shows dataset statistics grouped by whether the incident was extended or not
data.groupby("extended").describe()
```

```python
# Describes the distribution of incidents by country
plt.hist(data.country)
plt.show()
```

```python
# Prints the ammount of countries in which incidents occured
print(len(data.country.unique()))
```
